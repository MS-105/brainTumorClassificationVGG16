```python
#to know the file paths 
import os 
#for image 
from PIL import Image
```

```python
base_add = "../dataset/"
dir = ("Training/", "Testing/") 
classes = os.listdir("../dataset/Testing") #train and test has same number of class so we can take any one 
```

```python
#to count the number of images in each label 
for t in dir:
    print("in"+t)
    for label in classes:
        temp = base_add + t + label 
        count = len(os.listdir(temp))
        print(f"{label} : {count}")
```
